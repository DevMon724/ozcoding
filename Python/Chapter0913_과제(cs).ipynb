[문제1]

list와 tuple의 주된 차이점 은 무엇인가요?

정답:list 와 tuple의 주된 차이점은 가변성이다. 리스트 안의 값들은 추가, 수정, 삭제가 가능하지만 tuple의 값들은 변경이 불가능하다.
tuple의 값을 바꾸려면 리스트형으로 변환한 뒤 값을 변경 후 다시 튜플로 변환하는 방법이 있다.

[문제2]

Python의 set에 대해 설명해 주세요

정답:중복된 값을 가질 수 없는 자료형. 고유한 값만을 저장할 수 있고 순서(인덱스)가 없다.  

[문제3]
Python의 list는 연결리스트(linked list) 인가요?

정답: 파이썬에서 리스트는 동적인 자료구조를 가지고 있다. 연결 리스트는 각 노드가 데이터와 포인터를 가지고 한 줄로 연결되어 있는 방식으로 데이터를 참조하는 선형 자료 구조이다.
연결리스트는 데이터의 추가와 삭제가 자유롭지만 특정 요소에 접근하려면 순차적으로 접근해야하는 단점이 있고, 파이썬의 동적 배열은 데이터 추가, 삭제에 시간이 걸리는 단점이 있다.
